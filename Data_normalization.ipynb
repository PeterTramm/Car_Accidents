# 1 NF
<b>Rules of 1st Normal Form</b>
- Each table should contain a single value.
- Each record needs to be unique.

<b>Steps to take</b>
- Separate repeating groups into their own table -> No repeating groups


<b>Table form</b>

ACCIDENT_INFO('ACCIDENT_NO', 'ACCIDENT_DATE', 'ACCIDENT_TIME', 'ACCIDENT_TYPE','ACCIDENT_TYPE_DESC', 'DAY_WEEK_DESC', 'DCA_CODE', 'DCA_DESC','LIGHT_CONDITION', 'NODE_ID', 'NO_OF_VEHICLES', 'NO_PERSONS_KILLED','NO_PERSONS_INJ_2', 'NO_PERSONS_INJ_3', 'NO_PERSONS_NOT_INJ','NO_PERSONS', 'POLICE_ATTEND', 'ROAD_GEOMETRY', 'ROAD_GEOMETRY_DESC','SEVERITY', 'SPEED_ZONE', 'RMA')

<b>Check for partial dependencies </b>
- PK is a single-attribute, hence no partial dependecies
# 2NF 
<b>Rules of 2nd Normal Form</b>
- The table should be in 1NF.
- Primary Key does not functionally dependent on any subset of Candidate Key.
- If the relation has a composite PK, then each non-key attribute must be fully dependent on the entire PK and not on a subset of the PK (i.e., there must be no partial dependency or augmentation).

<b> Steps to take</b>
* Remove partial dependencies 


<b>Table form</b>

ACCIDENT_INFO('ACCIDENT_NO', 'ACCIDENT_DATE', 'ACCIDENT_TIME', 'ACCIDENT_TYPE','ACCIDENT_TYPE_DESC', 'DAY_WEEK_DESC', 'DCA_CODE', 'DCA_DESC','LIGHT_CONDITION', 'NODE_ID', 'NO_OF_VEHICLES', 'NO_PERSONS_KILLED','NO_PERSONS_INJ_2', 'NO_PERSONS_INJ_3', 'NO_PERSONS_NOT_INJ','NO_PERSONS', 'POLICE_ATTEND', 'ROAD_GEOMETRY', 'ROAD_GEOMETRY_DESC','SEVERITY', 'SPEED_ZONE', 'RMA')

<b> Check for Transitivte Dependents </b>

ACCIDENT_NO -> {ACCIDENT_TYPE} -> ACCIDENT_TYPE_DESC

ACCIDENT_NO -> {DCA_CODE}      -> DCA_DESC

ACCIDENT_NO -> {ROAD_GEOMETRY} -> ROAD_GEOMETRY_DESC

# 3NF

<b>Rules of 3rd Normal Form</b>
- The tables should be in 2NF.
- There will be no Transitive Dependency.


<b>Table form</b>

ACCIDENT_INFO(<b>'ACCIDENT_NO'</b>,'ACCIDENT_DATE','ACCIDENT_TIME','ACCIDENT_TYPE',DAY_WEEK_DESC','POLICE_ATTEND',
'NO_PERSONS','NO_OF_VEHICLES','LIGHT_CONDITION','NO_PERSONS_KILLED','NO_PERSONS_INJ_2','NO_PERSONS_INJ_3',
'NO_PERSONS_NOT_INJ','SEVERITY', 'SPEED_ZONE', 'RMA','NODE_ID','DCA_CODE','ROAD_GEOMETRY')

ACCIDENT_TYPE(<b>'ACCIDENT_TYPE'</b>,'ACCIDENT_TYPE_DESC',)

DCA_INFO(<b>'DCA_CODE'</b>, 'DCA_DESC',)

ROAD_GEOMETRY(<b>'ROAD_GEOMETRY'</b>, 'ROAD_GEOMETRY_DESC')

# 2NF 
<b>Rules of 2nd Normal Form</b>
- The table should be in 1NF.
- Primary Key does not functionally dependent on any subset of Candidate Key.
- If the relation has a composite PK, then each non-key attribute must be fully dependent on the entire PK and not on a subset of the PK (i.e., there must be no partial dependency or augmentation).

<b> Steps to take</b>

<b>Table form</b>

ACCIDENT_EVENT(<b>'ACCIDENT_NO', 'EVENT_SEQ_NO'</b>, 'EVENT_TYPE', 'EVENT_TYPE_DESC',

<b> Check for Transitivte Dependents </b>
{ACCIDENT-NO, EVENT_SEQ_NO} -> EVENT_TYPE -> EVENT_TYPE_DESC
{ACCIDENT-NO, EVENT_SEQ_NO} -> OBJECT_TYPE -> OBJECT_TYPE_DESC

# 3NF

<b>Rules of 3rd Normal Form</b>
- The tables should be in 2NF.
- There will be no Transitive Dependency.


<b>Table form</b>


ACCIDENT_EVENT(<b>'ACCIDENT_NO', 'EVENT_SEQ_NO'</b>, 'EVENT_TYPE','VEHICLE_1_ID', 'VEHICLE_1_COLL_PT', ,'VEHICLE_2_ID', 'VEHICLE 2 COLL PT DESC', 'PERSON_ID', 'OBJECT_TYPE')

EVENT_TYPE(<b>EVENT_TYPE</b>, EVENT_TYPE_DESC)

VEHICLE_COL1(<b>VEHICLE_1_COLL_PT<b>,VEHICLE 1 COLL DESC)

VEHICLE_COL1(<b>VEHICLE_2_COLL_PT<b>,VEHICLE 2 COLL DESC)






```python
Object_Type = accident_event_df[["OBJECT_TYPE","OBJECT_TYPE_DESC"]].drop_duplicates().reset_index(drop=True)
Event_Type = accident_event_df[["EVENT_TYPE","EVENT_TYPE_DESC"]].drop_duplicates().reset_index(drop=True)
Vehicle_Coll1 = accident_event_df[["VEHICLE_1_COLL_PT","VEHICLE 1 COLL PT DESC"]].drop_duplicates().reset_index(drop=True)

Vehicle_Coll2 = accident_event_df[["VEHICLE_2_COLL_PT","VEHICLE 2 COLL PT DESC"]].drop_duplicates().reset_index(drop=True)

accident_event_df = accident_event_df.drop(["OBJECT_TYPE_DESC","EVENT_TYPE_DESC","VEHICLE 2 COLL PT DESC","VEHICLE 1 COLL PT DESC"],axis=1)
```

```python
# Renaming column for concat later
Vehicle_Coll2= Vehicle_Coll2.rename(columns={"VEHICLE_2_COLL_PT":"VEHICLE_1_COLL_PT",
                                             "VEHICLE 2 COLL PT DESC":
                                             "VEHICLE 1 COLL PT DESC"})
# Combing both Coll PT tables into one table. 
Vehicle_Coll = pd.concat([Vehicle_Coll1,Vehicle_Coll2])
Vehicle_Coll = Vehicle_Coll.drop_duplicates().reset_index(drop=True).rename(columns={"VEHICLE_1_COLL_PT":"VEHICLE_COLL_PT",
                                                                                    "VEHICLE 1 COLL PT DESC":"VEHICLE COLL PT DESC"})
Vehicle_Coll
```
